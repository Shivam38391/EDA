```python
class Vehicle:
    def Vehicle_info(self):
        self.car_info()
        print('Inside Vehicle class')

    def car_info(self):
        print(f"inside Vehical car_info")

# Child class
class Car(Vehicle):
    def car_info(self):
        print('Inside Car over ridden car_info')

    def Vehicle_info(self):
        super().Vehicle_info()
        print("Inside Vehicle info of Car..")

# Create object of Car
car = Car()
car.Vehicle_info()
```
